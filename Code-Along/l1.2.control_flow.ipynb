```python
# Example - Oil leagage

# There is a oil leakage casing the bird population on an island to halv each year. Fromn the start there were 80000 bird, how many years does it take until only one tenth arr  remaining?

birds = 80000
year = 0 

while birds > 8000:
    print(f"Year {year}, {birds: 0f} birds")
    birds /= 2
    year += 1 

print (f"It takes {year} years for the birds to raech one tenth.")
```

```python
#tärning

import random as rng 
num_sixes = 0
dice_rolls = 100000000

for i in range(dice_rolls):
    dice = rng.randint(1, 6)
    if dice == 6:
        num_sixes += 1

print(f"Number of sixes {num_sixes} in {dice_rolls} dice rolls.")
print(f"Probability of six: {num_sixes / dice_rolls}")
```
