```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

college = pd.read_csv("College.csv")
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")
college = college3
college.head()
```

```python
fig, axs = plt.subplots(1,1)
sns.boxplot(data=college,x="Outstate",y='Private',ax=axs)
```

```python
fig, axs = plt.subplots(1,1)
sns.boxplot(data=college,x="Private",y='Outstate',ax=axs)
```

### (f)
Elite - Skolor som har ver 50% av elever som r top10pers
outstate - Avgiften fr skolan.   

```python
college["Elite"] = pd.cut(college["Top10perc"], [0,50,100], labels=["No","Yes"])
print(college.value_counts("Elite"))
fig, axs = plt.subplots(1,1)
sns.boxplot(data=college, x="Elite", y='Outstate', ax=axs)
```

```python
fig, ((enroll, grad_rate), (sf_ratio, phd)) = plt.subplots(2,2, figsize=(10,8))

enroll.hist(data=college, x="Enroll", label="Enroll", bins=10)
enroll.set_title("Enroll")
grad_rate.hist(data=college, x="Grad.Rate", label="Grad.Rate", bins=20)
grad_rate.set_title("Grad.Rate")
sf_ratio.hist(data=college, x="S.F.Ratio", bins=30)
sf_ratio.set_title("SF Ratio")
phd.hist(data=college, x="PhD",bins=50)
phd.set_title("Teachers with PhD")
```
