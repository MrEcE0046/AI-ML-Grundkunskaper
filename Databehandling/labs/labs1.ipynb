### (h)
Summering, personliga upptÃ¤ckter

Topp 15 grad rated college skolor och dess utgift per elev. 

```python
college["Grad.Rate"].describe()
college[college["Grad.Rate"] == 118] # Hittade ett fel i grad.rate.

highest_g_rate = college.sort_values("Grad.Rate")

highest_g_rate = highest_g_rate.head(776)
highest_g_rate = highest_g_rate.tail(15).reset_index()

fig, ax1 = plt.subplots()
ax1.barh(highest_g_rate["College"], highest_g_rate["Grad.Rate"], height=0.5, label="Grad.Rate", color="blue", alpha=0.5)
ax1.set_xlabel('Graduation Rate %')
ax1.set_xlim(0, 100)


ax2 = ax1.twiny()
ax2.barh(highest_g_rate["College"], highest_g_rate["Expend"], height=0.5, label="Expend", color="green")
ax2.set_xlabel('Expenditures $/student')
plt.title("College expends per student and graduation rate")
fig.legend(title="Graduation rate vs Expend", bbox_to_anchor=(0.9, 0.9), loc='upper left')


plt.show()
```

En Ã¶verblick pÃ¥ de 15 skolorna som har hÃ¶gst procentuell eximinationer och dess utgift per elev. Mest troligt har skolorna olika program och dÃ¤rfÃ¶r varierar utgifterna mycket.\
Det kÃ¤nns tryggt att de generella topskolorna som Harvard, Yale och Colombia har de hÃ¶gsta utgifterna men ocksÃ¥ imponerande att de har bland de hÃ¶gsta examensfrekvens. 

```python
elite = college.sort_values("Top10perc").tail()

top_elite_budget = elite[["Outstate","Expend","Personal"]]

top_elite_budget.plot(kind="barh", figsize=(15,4))
plt.title("Elite colleges cashflow")
plt.xlabel("$")
plt.tight_layout()
```

- Outstate : Out-of-state tuition
- Expend : Instructional expenditure per student
