```python
from labb3 import readfile, catagories,line
import matplotlib.pyplot as plt, numpy as np

def line1(x):
    return x * -0.489

def line2(x):
    return x * -2 + 0.16

def line3(x):
    return x * -1.08

def plott(label_data, x, y):
    """ Plot lines and scatterplotts """
    
    zeroes = [i for i in label_data if i[2] == 0]  
    ones = [i for i in label_data if i[2] == 1]
 
    plt.plot(x, y, "r", label=f"y={k}x+{m}")
    plt.plot(line1(x), x, "blue", label="f(x)=-0.489x")
    plt.plot(line2(x), x, "green", label="g(x)=-2x+0.16")
    plt.plot(line3(x), x, "orange", label="h(x)=-1.08x-0.1")
    
 
    plt.scatter(np.array(ones)[:, 0], np.array(ones)[:, 1], c="blue", label="Label 1")
    plt.scatter(np.array(zeroes)[:, 0], np.array(zeroes)[:, 1], c="green", label="Label 0")
    
    plt.title("Summary")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.legend()
    plt.grid()
    plt.ylim(-6,6)
    plt.xlim(-6, 5)
    plt.show()

k = -1.5
m = 0
data = readfile()
label_data = catagories(data, k, m)
x, y = line(k, m)
plott(label_data, x, y)
```
