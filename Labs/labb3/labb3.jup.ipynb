```python
from labb3 import readfile, catagories,line
import matplotlib.pyplot as plt, numpy as np

def line1(x):
    return x * -0.489

def line2(x):
    return x * -2 + 0.16

def line3(x):
    return x * -1.08


def plott(label_data, x, y, k, m):
    """ Plot a line and scatterplot """
    
    label_data=np.array(label_data,dtype=float)
    a = label_data[:, 0]
    b = label_data[:, 1]
    label = label_data[:, 2]

    plt.plot(x, y, "r", label=f"y={k}x+{m}")
    plt.plot(line1(x), x, "blue", label="f(x)=-0.489x")
    plt.plot(line2(x), x, "green", label="g(x)=-2x+0.16")
    plt.plot(line3(x), x, "orange", label="h(x)=-1.08x-0.1")

    plt.scatter(a[label==0],b[label==0], c="blue", label="Label 0")
    plt.scatter(a[label==1],b[label==1], c="red", label="Label 1")
    plt.title(f"{np.sum(label==0)} of label-0 | {np.sum(label==1)} of label-1")
    plt.xlabel("X-axle")
    plt.ylabel("Y-axle")
    plt.legend()
    plt.grid()
    plt.axis([-6,5,-6,6])
    plt.show()


k = -1.5
m = 0
data = readfile()
label_data = catagories(data, k, m)
x, y = line(k, m)
plott(label_data, x, y, k, m)
```
