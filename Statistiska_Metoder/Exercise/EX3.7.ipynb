```python
# 8.
import pandas as pd
import numpy as np
import statsmodels.api as sm

auto_df = pd.read_csv("Auto.csv")

# Tar bort alla sankande värden
auto_df = auto_df.drop(auto_df[auto_df.values == "?"].index)
auto_df = auto_df.reset_index()
auto_df.head(10)
```

```python
# Konverterar kvantitiva variabler till numeriska. 

datatype = {"quant": ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "year"], 
            "qual": ["origin", "name"]}

quants = auto_df[datatype["quant"]].astype(np.float_)
auto_df = pd.concat([quants, auto_df[datatype["qual"]]], axis=1)
```

```python
# börjar med att implementera linjär regression from början med hjälp av numpy.

intercept_const = pd.DataFrame({"intercept": np.ones(auto_df.shape[0])})

X = pd.concat([intercept_const, auto_df["horsepower"]], axis=1)
y = auto_df["mpg"]

def linear_model(X, y):
    # Estimation av parametrar för den linjära regressions modellen med OLS, ordinary least squares. 
    XTX_inv = np.linalg.inv(X.T @ X)
    XTy = X.T @ y
    beta = XTX_inv @ XTy
    return beta

def predict(beta, X):
    # Förutspår y given beta parametrar och X variabler
    return X @ beta

beta = linear_model(X, y)
y_pred = predict(beta, X)
```
