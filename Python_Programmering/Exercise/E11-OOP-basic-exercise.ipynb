````python
class Convert:
    """An class to convert us to western modern units."""
    def __init__(self, x):
        if not isinstance(x, (float, int)):
                raise TypeError(f"{x} is not a valid number.")
        elif len(x) <= 0:
            raise ValueError("Vector can't be empty.")
        self.x = x
    
    @property
    def inch_to_cm(self):
        return Convert(self.x*2.54)
    @property
    def foot_to_meter(self):
        return Convert(self.x*0.3048)
    @property
    def pound_to_kg(self):
        return Convert(self.x*0.45)

    def __repr__(self):
        return f"{self.x}"
    
unit = Convert(5)
print(f"5 inch is {unit.inch_to_cm}cm")
print(f"5 foot is {unit.foot_to_meter}m")
print(f"5 pound is {unit.pound_to_kg}kg")
````
